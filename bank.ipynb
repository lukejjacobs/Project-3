```python
df = pd.read_csv(Path("./Resources/Customer-Churn-Records.csv"))
```

```python
age_bins = [0,19.9,29.9,39.9,49.9,59.9,69.9,79.9,120]
age_groups = ["<20s", "20s", "30s", "40s", "50s", "60s", "70s", ">80s"]
```

```python
credit_bins = [0,399.9,499.9,599.99,699.99,799.99,850]
credit_groups = ["<400", "400s", "500s", "600s", "700s", "<800"]
```

```python
age_categories = pd.cut(df["Age"], bins = age_bins, labels = age_groups)
df["AgeCategories"] = age_categories
credit_categories = pd.cut(df["CreditScore"], bins = credit_bins, labels = credit_groups)
df["CreditScoreCategories"] = credit_categories
```
